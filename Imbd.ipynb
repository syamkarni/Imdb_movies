```python

#r=requests.get('https://www.imdb.com/chart/top/')

# Define a user agent
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
# Make the request with the defined user agent
r = requests.get('https://www.imdb.com/chart/top/', headers=headers)
soup=bs(r.content)
contents=soup.prettify()
print(contents)
```

create empty lists

```python
movie_title=[]
movie_year=[]
movie_rating=[]
```

Extract html tag contents

```python
imbd_table = soup.find(class_='chart_full_width')
```

```python
movie_title=soup.find_all(class_='ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-be6f1408-9 srahg cli-title')
movie_title = [title.find('h3').text.strip().split('. ')[1] for title in movie_title]
movie_title
```

```python
movie_rating=soup.find_all(class_='sc-e2dbc1a3-0 ajrIH sc-be6f1408-2 dAeZAQ cli-ratings-container')
movie_rating=[rating.find('span').text.strip()[:3] for rating in movie_rating]
movie_rating
```

```python
movie_year=soup.find_all(class_='sc-be6f1408-7 iUtHEN cli-title-metadata')
movie_year=[year.find('span').text.strip() for year in movie_year]
movie_year
```

Create a Dataframe from the obtained data

```python
movie_df=pd.DataFrame({'Movie Title:': movie_title, 'Year:': movie_year, 'Rating:': movie_rating})
movie_df.head(30)
```
